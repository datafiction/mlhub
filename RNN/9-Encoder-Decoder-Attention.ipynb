````python
import numpy as np
from keras.preprocessing import sequence
from keras.models import Sequential
from keras.layers import Dense, Dropout, Embedding,\
LSTM, Bidirectional
from keras.datasets import imdb
````

````python
import random as random
import numpy as np
from matplotlib import pyplot
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
def get_sequence(length):
    freq = 5
    noise = [random.uniform(-0.3,0.3) for i in range(length)]
    sequence = [np.sin(2 * np.pi * freq * (i/length)) for i in range(length)]
    signal = np.array(noise) + np.array(sequence)
    return signal
````

````python
# Plot sequence
sequence = get_sequence(100)
plt.figure(figsize = [15,3])
plt.plot(sequence)
plt.show()
````

````python
def generate_examples(length, n_samples, output):
    
    X, y = list(), list()
    
    for k in range(n_samples):
        
        sequence1 = get_sequence(length)
        sequence2 = get_sequence(output)
        
        # input output both will have same size
        X.append(sequence1)
        y.append(sequence2)
        
    X = np.array(X).reshape(n_samples, length, 1)
    y = np.array(y).reshape(n_samples, output, 1)
    
    return X, y
````

````python
class Encoder(tf.keras.layers.Layer):
    
      def __init__(self,sequence_length, output_length):
            
        super(Encoder, self).__init__()
        

        # The encoder bi-directional lstm
        self.bilstm = tf.keras.layers.Bidirectional(LSTM(50,return_sequences=True,\
                       input_shape=(sequence_length, 1),name='input-layer'))

        # The GRU RNN layer processes those vectors sequentially.
        # Return the sequence and state
        self.incoder_lstm = tf.keras.layers.LSTM(50,return_sequences=True,\
                                       return_state=True,\
                                       recurrent_initializer='glorot_uniform')
````

````python
class BahdanauAttention(tf.keras.layers.Layer):
    
    def __init__(self, units):
            super().__init__()
            # For Eqn. (4), the  Bahdanau attention
            self.W1 = tf.keras.layers.Dense(units, use_bias=False)
            self.W2 = tf.keras.layers.Dense(units, use_bias=False)

            self.attention = tf.keras.layers.AdditiveAttention()

    def call(self, query, value, mask):
            

            # From Eqn. (4), `W1@ht`.
            w1_query = self.W1(query)
            
            # From Eqn. (4), `W2@hs`.
            w2_key = self.W2(value)
            

            query_mask = tf.ones(tf.shape(query)[:-1], dtype=bool)
            value_mask = mask

            context_vector, attention_weights = self.attention(
                inputs = [w1_query, value, w2_key],
                mask=[query_mask, value_mask],
                return_attention_scores = True,
            )

            return context_vector, attention_weights
````

````python
class Decoder(tf.keras.layers.Layer):
    
      def __init__(self, sequence_length, output_length):
            super(Decoder, self).__init__()
            

            # The decoder bi-directional lstm
            self.bilstm = tf.keras.layers.Bidirectional(LSTM(50,return_sequences=True,\
                            input_shape=(sequence_length, 1),name='input-layer'))

            # The middle layer lstm
            self.decoder_lstm = tf.keras.layers.LSTM(50,return_sequences=True,\
                                       return_state=True,\
                                       recurrent_initializer='glorot_uniform')

            # For step 3. The RNN output will be the query for the attention layer.
            self.attention = BahdanauAttention(sequence_length)

            # For step 4. Eqn. (3): converting `ct` to `at`
            self.Wc = tf.keras.layers.Dense(sequence_length,\
                                            activation=tf.math.tanh,\
                                            use_bias=False)

            # For step 5. This fully connected layer produces the logits for each
            # output token.
            self.fc = tf.keras.layers.Dense(100)
````
